```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# %matplotlib inline 
# %config InlineBackend.figure_format='retina'
```

```python
import pandas as pd
elec = pd.read_csv('./data/1인당 전력소비량(1990, 2000, 2010~2019).csv')
gas = pd.read_csv('./data/온실가스(1990~2019).csv')
```

```python
gas.set_index('분류', inplace=True)
elec.set_index('분류', inplace=True)
```

```python
display(gas)
display(elec)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10,6))
plt.title('온실가스와 전력소비량 증가 추이', fontsize=20)
plt.xlabel('연도')

plt.plot(gas.iloc[0,21:]/100, 'o-')#, label='온실가스(100Gg CO2eq.)')
plt.plot(elec.iloc[0,2:], 'o-')#, label='전력소비량(kWh)')

plt.legend(['온실가스(100Gg CO2eq.)', '전력소비량(kWh)'])
plt.savefig("output/온실가스와 전력소비량 추이.png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))
plt.title('온실가스 증가 추이')
plt.plot(gas.iloc[0,], color='green', label='온실가스')
plt.legend()
plt.xticks(rotation='70')
plt.savefig("output/온실가스 증가 추이.png")

plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot()

ax.plot(gas.iloc[0,21:]/100, color='green', label='온실가스')
ax.set_ylabel('온실가스', fontsize=15)

ax2=ax.twinx()
ax2.plot(elec.iloc[0,2:], color='orange', label='전력소비량')
ax2.set_ylabel('전력소비량', fontsize=15)


plt.title('온실가스와 전력소비량 증가 추이', fontsize=20)

ax.legend()
ax2.legend(loc='upper right')
plt.savefig("output/온실가스와 전력소비량 추이.png")

plt.show()
```

```python
import pandas as pd
temp = pd.read_excel('./data/연평균기온.xls', skiprows = 2, skipfooter = 1)
temp.rename(columns={'Unnamed: 0':'평균'}, inplace=True)
temp.set_index('평균', inplace=True)
temp
```

```python
temp.iloc[0,]
```

```python
gas = pd.read_csv('./data/온실가스(1990~2019).csv')
gas.iloc[:,22:]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot()

ax.plot(gas.iloc[0,22:]/100, color='green', label='온실가스')
ax.set_ylabel('온실가스', fontsize=15)

ax2=ax.twinx()
ax2.plot(temp.iloc[0,], color='orange', label='연평균 기온')
ax2.set_ylabel('연평균 기온', fontsize=15)
ax2.set_ylim(10.0, 15.0)

plt.title('온실가스와 연평균 기온 증가 추이', fontsize=20)

ax.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.savefig("output/온실가스와 연평균 기온 증가 추이.png")

plt.show()
```
