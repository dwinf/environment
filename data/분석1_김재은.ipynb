```python
from matplotlib import font_manager, rc
font_path = "data/malgun.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import scipy.stats as stats
import numpy as np
```

```python
df = pd.read_csv('data/서울시 평균기온.csv', index_col=0)
display(df)
```

```python
df = pd.read_csv('data/전력사용량과 평균기온.csv', index_col=0)
display(df)
df.corr(method = 'pearson')
```

```python
sns.pairplot(df[['월평균기온', '전체 전력 사용량 (MWh)']], kind='reg')
```

```python
df[['월평균기온', '전체 전력 사용량 (MWh)']].corr(method = 'pearson')
```

```python
x = df['월평균기온'].values.tolist()
y = df['전체 전력 사용량 (MWh)'].values.tolist()
```

```python
stats.pearsonr(x, y)
```

```python
df = pd.read_csv('data/연평균기온+1인당전력소비량.csv', index_col=0)
display(df)
```

```python
df.T.corr(method = 'pearson')
```

```python
stats.pearsonr(df.loc['연평균 기온'], df.loc['인구 1인당 전력소비량 (kWh/인)'])
```

```python
sns.pairplot(df.T, kind='reg')
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot()
ax.plot(df.loc['연평균 기온'], color='green')
ax.legend(['연평균 기온 (℃)'], loc='upper left')
ax2 = ax.twinx()
ax2.plot(df.loc['인구 1인당 전력소비량 (kWh/인)'], color='orange')
ax2.legend(['인구 1인당 전력소비량 (kWh/인)'], loc='upper right')
```
